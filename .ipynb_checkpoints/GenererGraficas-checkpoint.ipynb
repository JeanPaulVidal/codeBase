```python
import numpy as np
import matplotlib.pyplot as plt
from prettytable import PrettyTable

def frec_abs(datos_entrada):
    # Convertir los datos a mayúsculas y eliminar espacios en blanco
    datos_entrada = [str(x).upper().strip() for x in datos_entrada]
    
    # Calcular la frecuencia absoluta
    _, conteos = np.unique(datos_entrada, return_counts=True)
    frecuencias_absolutas = dict(zip(np.unique(datos_entrada), conteos))
    
    return frecuencias_absolutas

def frec_rel(frecuencias_absolutas):
    total = sum(frecuencias_absolutas.values())
    frecuencias_relativas = {k: v/total for k, v in frecuencias_absolutas.items()}
    return frecuencias_relativas

def frec_acum(frecuencias_relativas):
    frecuencias_acumuladas = {}
    acumulado = 0
    for clase, frecuencia in sorted(frecuencias_relativas.items()):
        acumulado += frecuencia
        frecuencias_acumuladas[clase] = acumulado
    return frecuencias_acumuladas

def generar_graficas(datos):
    # Calcular las frecuencias absolutas
    frecuencias_absolutas = frec_abs(datos)
    
    # Calcular las frecuencias relativas
    frecuencias_relativas = frec_rel(frecuencias_absolutas)
    
    # Calcular las frecuencias acumuladas
    frecuencias_acumuladas = frec_acum(frecuencias_relativas)
    
    # Obtener las clases (límites inferiores y superiores)
    clases_sorted = sorted(list(set(datos)))
    
    # Generar los límites inferiores y superiores
    limites_inferiores = {clase: clase - 0.165 for clase in clases_sorted}
    limites_superiores = {clase: clase + 0.165 for clase in clases_sorted}
    
    # Calcular las marcas centrales
    marcas_centrales = {clase: (limites_inferiores[clase] + limites_superiores[clase]) / 2 for clase in clases_sorted}
    
    # Generar los gráficos
    fig, axes = plt.subplots(2, 2, figsize=(12, 9))
    
    # Histograma
    axes[0, 0].hist(datos, bins=len(clases_sorted), edgecolor='black')
    axes[0, 0].set_title('Histograma')
    axes[0, 0].set_xlabel('Valor')
    axes[0, 0].set_ylabel('Frecuencia')
    
    # Polígono de frecuencias
    axes[0, 1].plot(list(marcas_centrales.values()), list(frecuencias_absolutas.values()), marker='o')
    axes[0, 1].set_title('Polígono de frecuencias')
    axes[0, 1].set_xlabel('Valor')
    axes[0, 1].set_ylabel('Frecuencia')
    
    # Ojiva
    axes[1, 0].plot(sorted(clases_sorted), list(frecuencias_acumuladas.values()))
    axes[1, 0].set_title('Ojiva')
    axes[1, 0].set_xlabel('Valor')
    axes[1, 0].set_ylabel('Frecuencia acumulada')
    
    # Gráfico de pastel
    axes[1, 1].pie(list(frecuencias_absolutas.values()), labels=list(frecuencias_absolutas.keys()), autopct='%1.1f%%')
    axes[1, 1].set_title('Gráfico de pastel')
    axes[1, 1].axis('equal')
    
    plt.subplots_adjust(hspace=0.5, wspace=0.5)
    plt.show()
    
    # Imprimir los resultados en una tabla
    imprimir_tabla(datos, frecuencias_absolutas, frecuencias_relativas, frecuencias_acumuladas, limites_inferiores, limites_superiores, marcas_centrales)

def tabla_agrupada(datos, frecuencias_absolutas, frecuencias_relativas, frecuencias_acumuladas, limites_inferiores, limites_superiores, marcas_centrales):
    tabla = PrettyTable()
    tabla.field_names = ["Clase", "Frecuencia Absoluta", "Frecuencia Relativa", "Frecuencia Acumulada", "Límite Inferior", "Límite Superior", "Marca Central"]
    
    for clase in sorted(set(datos)):
        frecuencia_absoluta = frecuencias_absolutas.get(str(clase), 0)
        frecuencia_relativa = frecuencias_relativas.get(str(clase), 0)
        frecuencia_acumulada = frecuencias_acumuladas.get(str(clase), 0)
        limite_inferior = limites_inferiores.get(clase, "")
        limite_superior = limites_superiores.get(clase, "")
        marca_central = marcas_centrales.get(clase, "")
        tabla.add_row([clase, frecuencia_absoluta, "{:.2f}".format(frecuencia_relativa), "{:.2f}".format(frecuencia_acumulada), "{:.2f}".format(limite_inferior), "{:.2f}".format(limite_superior), "{:.2f}".format(marca_central)])
    
    print(tabla)

```

```python
# Ejemplo de uso
datos = [0, 0, 0.33, 0.33, 0.33, 0.33, 0.66, 0.66, 0.66, 0.66, 0.99, 0.99, 0.99, 1.32, 1.32, 1.32, 1.65, 1.65, 1.65, 1.65]
generar_graficas(datos)
```
